```python
import pandas as pd

import csv
def apriori(transactions, min_support):
    # Count the occurrences of each item in the dataset
    item_counts = {}
    for transaction in transactions:
        for item in transaction:
            item_counts[item] = item_counts.get(item, 0) + 1

    # Remove items that do not meet the minimum support threshold
    items = set(item for item, count in item_counts.items() if count >= min_support)

    # Generate frequent itemsets of size 1
    frequent_itemsets = {frozenset((item,)) for item in items}

    # Generate frequent itemsets of larger sizes
    k = 2
    while True:
        # Generate candidate itemsets of size k
        candidate_itemsets = set()
        for itemset1 in frequent_itemsets:
            for itemset2 in frequent_itemsets:
                if len(set(itemset1).union(set(itemset2))) == k:
                    candidate_itemsets.add(tuple(sorted(set(itemset1).union(set(itemset2)))))
        print("Candidate itemsets:", candidate_itemsets)
        print("")

        # Count the occurrences of each candidate itemset in the dataset
        itemset_counts = {itemset: 0 for itemset in candidate_itemsets}
        for transaction in transactions:
            for itemset in candidate_itemsets:
                if set(itemset).issubset(set(transaction)):
                    itemset_counts[itemset] += 1

        # Remove itemsets that do not meet the minimum support threshold
        frequent_itemsets = {itemset for itemset, count in itemset_counts.items() if count >= min_support}

        # If no frequent itemsets of size k are found, terminate
        if not frequent_itemsets:
            break

        k += 1

    # Plot the itemset support distribution
    itemset_supports = sorted([count/len(transactions) for itemset, count in itemset_counts.items()])

    return frequent_itemsets

# Define the transactional dataset
transactions = []
with open('datset1.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        transaction = set(row)
        transactions.append(transaction)

# Generate frequent itemsets with a minimum support of 3
frequent_itemsets = apriori(transactions, min_support=3)
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel('datset.xlsx')
df = df.apply(lambda x: pd.Series(x.dropna().values)).fillna('')
frequent_itemsets_apriori = apriori(df.astype('bool'), min_support=0.4, use_colnames=True)
rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=0.45)
print("Results using Apriori Algorithm")
print(frequent_itemsets_apriori)
print(rules_apriori)
```
