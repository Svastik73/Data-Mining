```python
import pandas as pd
import pyfpgrowth as p
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

dataset= pd.read_excel("datset.xlsx",na_filter=False)
transactions = []
for i in range(0, dataset.shape[0]):
    transaction=[]
    for j in range(0, dataset.shape[1]):
        if dataset.values[i,j]!='':
           transaction.append(dataset.values[i,j])
    transactions.append(transaction)
patterns = p.find_frequent_patterns(transactions,0.40)
print(patterns)
for items,cnt in patterns.items():
    print(f'{items}-->{cnt}')
rules = p.generate_association_rules(patterns,0.45)
for items,cnt in rules.items():
    print(f'{items}-->{cnt}')
frequent_itemsets_fpgrowth = fpgrowth(dataset.astype('bool'), min_support=0.4)
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=0.45)
print("Results using F-P Growth Algorithm")
print(frequent_itemsets_fpgrowth)
print(rules_fpgrowth)
```
